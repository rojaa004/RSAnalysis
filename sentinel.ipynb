```python
# # Earth Engine Satellite Data Analysis and Visualization
#
# This notebook processes Sentinel-2 and Landsat 8 satellite imagery to calculate
# Vegetation (NDVI) and Water (NDWI) indices, visualizes these indices and
# farm boundaries on an interactive map, calculates statistics for the indices
# and elevation within each farm polygon, and exports a Digital Elevation Model (DEM)
# for the area.
#

# # Setup and Initialization
#
# This section sets up the necessary libraries, authenticates and initializes
# Earth Engine, and defines common parameters like the time period and farm boundaries.
#

import ee
import folium
import pandas as pd
from google.colab import files # Required for downloading files in Colab

# Authenticate Earth Engine. This will prompt you to authorize access.
# ee.Authenticate() # Assuming this is done in a previous cell and is not needed here.

# Initialize Earth Engine with your project ID.
try:
    ee.Initialize(project='sentinel-project-ndvi')
    print("Earth Engine initialized successfully.")
except Exception as e:
    print(f"Earth Engine initialization failed: {e}")
    print("Please make sure you have authenticated and specified your project ID if required.")


# Define a time period for filtering image collections
start_date = '2024-01-01'
end_date = '2024-12-31'

# Define farm boundaries using a FeatureCollection asset.
# This variable is used for both Sentinel and Landsat clipping and visualization.
# Replace with the correct asset path if different.
farmPolygons = ee.FeatureCollection("projects/sentinel-project-ndvi/assets/farm_polygons")

# Check if farm polygons were loaded successfully
if farmPolygons.size().getInfo() == 0:
    print("Error: Farm polygons could not be loaded or are empty. Please check the asset path.")
    # Exit or handle the error appropriately if polygons are essential.


# # Data Masking and Index Calculation Functions
#
# This section defines functions used to mask clouds in satellite images
# and calculate NDVI and NDWI.
#

# Define the maskS2clouds function for Sentinel-2
def maskS2clouds(image):
  """Masks cloudy pixels in a Sentinel-2 image.

  Args:
    image: An ee.Image object representing a Sentinel-2 image.

  Returns:
    An ee.Image object with cloudy pixels masked.
  """
  # Select the Quality Assessment band (QA60) for cloud information.
  qa = image.select('QA60')

  # Bits 10 and 11 are clouds and cirrus, respectively, in the QA60 band.
  cloudBitMask = 1 << 10
  cirrusBitMask = 1 << 11

  # Both flags should be set to zero, indicating clear conditions.
  # Create a mask where both cloud and cirrus bits are not set.
  mask = qa.bitwiseAnd(cloudBitMask).eq(0).And(
      qa.bitwiseAnd(cirrusBitMask).eq(0))

  # Return the masked image, with the mask applied to all bands.
  return image.updateMask(mask)


# Define the maskL8clouds function for Landsat 8 Collection 2 Tier 1 Surface Reflectance
def maskL8clouds(image):
    """Masks cloudy pixels and applies scaling factors in a Landsat 8 image.

    Args:
      image: An ee.Image object representing a Landsat 8 image.

    Returns:
      An ee.Image object with cloudy pixels masked and optical bands scaled.
    """
    # Select the Quality Assessment band.
    qa = image.select('QA_PIXEL')

    # Define bitmasks for cloud and cloud shadow based on the QA_PIXEL band documentation.
    cloud = qa.bitwiseAnd(1 << 3) # Bit 3 is cloud
    cloud_shadow = qa.bitwiseAnd(1 << 4) # Bit 4 is cloud shadow

    # Both flags should be set to zero, indicating clear conditions.
    # Create a mask where both cloud and cloud shadow bits are not set.
    mask = cloud.eq(0).And(cloud_shadow.eq(0))

    # Apply the scaling factors to the optical bands (B1-B7, B10, B11) as per Landsat Collection 2 documentation.
    # These bands need to be scaled for surface reflectance and brightness temperature.
    optical_bands = ['SR_B1', 'SR_B2', 'SR_B3', 'SR_B4', 'SR_B5', 'SR_B6', 'SR_B7']
    scaled_image = image.select(optical_bands).multiply(0.0000275).add(-0.2)
    # Note: Thermal bands (B10, B11) also require scaling for brightness temperature,
    # but they are not used in NDVI or NDWI calculations for this script.

    # Combine scaled bands with other bands (like QA_PIXEL) and apply the mask.
    # addBands(srcImg, names, overwrite) adds bands from srcImg to the current image.
    # None for names means use the names from srcImg. True for overwrite means replace existing bands with the same name.
    return image.addBands(scaled_image, None, True).updateMask(mask)


# Define a function to calculate NDWI for Sentinel-2
def calculate_sentinel_ndwi(image):
  """Calculates NDWI for a Sentinel-2 image.

  Args:
    image: An ee.Image object representing a Sentinel-2 image.

  Returns:
    An ee.Image object with an 'NDWI' band.
  """
  # Calculate NDWI using Green (B3) and NIR (B8) bands (Gao, 1996).
  ndwi = image.normalizedDifference(['B3', 'B8']).rename('NDWI')
  return image.addBands(ndwi)

# Define a function to calculate NDWI for Landsat 8
def calculate_landsat_ndwi(image):
    """Calculates NDWI for a Landsat 8 image.

    Args:
      image: An ee.Image object representing a Landsat 8 image.

    Returns:
      An ee.Image object with an 'NDWI' band.
    """
    # Calculate NDWI using Green (SR_B3) and NIR (SR_B5) bands (Gao, 1996).
    ndwi = image.normalizedDifference(['SR_B3', 'SR_B5']).rename('NDWI')
    return image.addBands(ndwi)


# # Satellite Data Processing and Map Visualization
#
# This section processes Sentinel-2 and Landsat imagery, calculates NDVI and NDWI,
# creates median composites for statistics, clips the first available image for visualization,
# and prepares layers for the interactive Folium map.
#

# --- Sentinel-2 Data Processing ---
# Access and filter Sentinel-2 image collection
sentinel_collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
    .filterDate(start_date, end_date) \
    .filterBounds(farmPolygons)

# Apply the cloud masking function
masked_sentinel_collection = sentinel_collection.map(maskS2clouds)

# Compute NDVI for Sentinel-2.
sentinel_ndvi_collection = masked_sentinel_collection.map(
    lambda image: image.normalizedDifference(['B8', 'B4']).rename('NDVI')
)

# Compute NDWI for Sentinel-2.
sentinel_ndwi_collection = masked_sentinel_collection.map(calculate_sentinel_ndwi)

# Compute median composites for statistics
sentinel_ndvi_composite = None
if sentinel_ndvi_collection.size().getInfo() > 0:
    sentinel_ndvi_composite = sentinel_ndvi_collection.median()

sentinel_ndwi_composite = None
if sentinel_ndwi_collection.size().getInfo() > 0:
    sentinel_ndwi_composite = sentinel_ndwi_collection.median()


# Select the first Sentinel NDVI image for visualization and clipping.
sentinel_ndvi_image = sentinel_ndvi_collection.first()
clipped_sentinel_ndvi_viz = None # Use a different variable name for visualization
if sentinel_ndvi_image is not None:
    clipped_sentinel_ndvi_viz = sentinel_ndvi_image.clip(farmPolygons)

# Select the first Sentinel NDWI image for visualization and clipping
sentinel_ndwi_image = sentinel_ndwi_collection.select('NDWI').first()
clipped_sentinel_ndwi_viz = None # Use a different variable name for visualization
if sentinel_ndwi_image is not None:
    clipped_sentinel_ndwi_viz = sentinel_ndwi_image.clip(farmPolygons)


# Define visualization parameters for Sentinel NDVI.
sentinel_ndvi_vis = {
    'min': 0.0,
    'max': 1.0,
    'palette': ['red', 'yellow', 'green'] # Common color palette for NDVI
}

# Define visualization parameters for NDWI
ndwi_vis = {
    'min': 0.0,
    'max': 1.0,
    'palette': ['00FFFF', '0000FF'] # Cyan to blue color palette for water
}


# --- Landsat Data Processing ---
# Define a Landsat collection (e.g., Landsat 8 Collection 2 Tier 1 Surface Reflectance)
landsat_collection = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
    .filterDate(start_date, end_date) \
    .filterBounds(farmPolygons)

# Apply the cloud masking function to the Landsat collection.
masked_landsat_collection = landsat_collection.map(maskL8clouds)

# Compute NDVI for Landsat.
landsat_ndvi_collection = masked_landsat_collection.map(
    lambda image: image.normalizedDifference(['SR_B5', 'SR_B4']).rename('NDVI')
)

# Compute NDWI for Landsat.
landsat_ndwi_collection = masked_landsat_collection.map(calculate_landsat_ndwi)

# Compute median composites for statistics
landsat_ndvi_composite = None
if landsat_ndvi_collection.size().getInfo() > 0:
    landsat_ndvi_composite = landsat_ndvi_collection.median()

landsat_ndwi_composite = None
if landsat_ndwi_collection.size().getInfo() > 0:
    landsat_ndwi_composite = landsat_ndwi_collection.median()


# Select the first Landsat NDVI image for visualization and clipping.
landsat_ndvi_image = landsat_ndvi_collection.first()
clipped_landsat_ndvi_viz = None # Use a different variable name for visualization
if landsat_ndvi_image is not None:
    clipped_landsat_ndvi_viz = landsat_ndvi_image.clip(farmPolygons)

# Select the first Landsat NDWI image for visualization and clipping
landsat_ndwi_image = landsat_ndwi_collection.select('NDWI').first()
clipped_landsat_ndwi_viz = None # Use a different variable name for visualization
if landsat_ndwi_image is not None:
    clipped_landsat_ndwi_viz = landsat_ndwi_image.clip(farmPolygons)


# Define visualization parameters for Landsat NDVI (can be the same as Sentinel NDVI).
landsat_ndvi_vis = {
    'min': 0.0,
    'max': 1.0,
    'palette': ['red', 'yellow', 'green'] # Using the same color palette as Sentinel NDVI
}

# Define visualization parameters for farm polygons.
farm_vis = {
    'color': 'FF0000',  # Red color for boundaries
    'fillColor': '00000000', # Transparent fill
    'width': 2 # Boundary line width
}


# --- Map Visualization ---
# Get the map ID and token for the clipped Sentinel NDVI layer.
sentinel_ndvi_tile_url = None
if clipped_sentinel_ndvi_viz is not None: # Use the viz variable
    try:
        sentinel_ndvi_map_id = clipped_sentinel_ndvi_viz.getMapId(sentinel_ndvi_vis)
        sentinel_ndvi_tile_url = sentinel_ndvi_map_id['tile_fetcher'].url_format
    except Exception as e:
        # Error getting map ID
        pass

# Get the map ID and token for the clipped Sentinel NDWI layer.
sentinel_ndwi_tile_url = None
if clipped_sentinel_ndwi_viz is not None: # Use the viz variable
    try:
        sentinel_ndwi_map_id = clipped_sentinel_ndwi_viz.getMapId(ndwi_vis)
        sentinel_ndwi_tile_url = sentinel_ndwi_map_id['tile_fetcher'].url_format
    except Exception as e:
        # Error getting map ID
        pass


# Get the map ID and token for the clipped Landsat NDVI layer.
landsat_ndvi_tile_url = None
if clipped_landsat_ndvi_viz is not None: # Use the viz variable
    try:
        landsat_ndvi_map_id = clipped_landsat_ndvi_viz.getMapId(landsat_ndvi_vis)
        landsat_ndvi_tile_url = landsat_ndvi_map_id['tile_fetcher'].url_format
    except Exception as e:
        # Error getting map ID
        pass

# Get the map ID and token for the clipped Landsat NDWI layer.
landsat_ndwi_tile_url = None
if clipped_landsat_ndwi_viz is not None: # Use the viz variable
    try:
        landsat_ndwi_map_id = clipped_landsat_ndwi_viz.getMapId(ndwi_vis)
        landsat_ndwi_tile_url = landsat_ndwi_map_id['tile_fetcher'].url_format
    except Exception as e:
        # Error getting map ID
        pass


# Get the map ID and token for the farm boundaries layer.
farm_tile_url = None
if farmPolygons is not None:
    try:
        farm_map_id = farmPolygons.getMapId(farm_vis)
        farm_tile_url = farm_map_id['tile_fetcher'].url_format
    except Exception as e:
        # Error getting map ID
        pass


# Create a Folium map object.
# Attempt to center the map on the farm polygons if available, otherwise use a default location.
Map = folium.Map(location=[0, 0], zoom_start=2) # Initialize with a default
if farmPolygons is not None:
    try:
        center = farmPolygons.first().geometry().centroid().coordinates().getInfo()
        lat, lon = center[1], center[0]
        Map = folium.Map(location=[lat, lon], zoom_start=12)
    except Exception as e:
        # Keep this print as it provides useful fallback information
        print(f"Could not center map on farm polygons: {e}. Using default location.")


# Add the clipped Sentinel NDVI layer to the map if available.
if sentinel_ndvi_tile_url:
    folium.TileLayer(
        tiles=sentinel_ndvi_tile_url,
        attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
        overlay=True, # Add as an overlay so other layers can be seen
        name='Sentinel NDVI over Farms' # Descriptive layer name for the control
    ).add_to(Map)

# Add the clipped Sentinel NDWI layer to the map if available.
if sentinel_ndwi_tile_url:
    folium.TileLayer(
        tiles=sentinel_ndwi_tile_url,
        attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
        overlay=True, # Add as an overlay so other layers can be seen
        name='Sentinel NDWI over Farms' # Descriptive layer name for the control
    ).add_to(Map)

# Add the clipped Landsat NDVI layer to the map if available.
if landsat_ndvi_tile_url:
     folium.TileLayer(
        tiles=landsat_ndvi_tile_url,
        attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
        overlay=True, # Add as an overlay
        name='Landsat NDVI over Farms' # Use a different name for the Landsat layer
    ).add_to(Map)

# Add the clipped Landsat NDWI layer to the map if available.
if landsat_ndwi_tile_url:
     folium.TileLayer(
        tiles=landsat_ndwi_tile_url,
        attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
        overlay=True, # Add as an overlay
        name='Landsat NDWI over Farms' # Use a different name for the Landsat layer
    ).add_to(Map)


# Add the farm boundaries layer to the map if available.
if farm_tile_url:
     folium.TileLayer(
        tiles=farm_tile_url,
        attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
        overlay=True, # Add as an overlay
        name='Farm Boundaries' # Descriptive layer name
    ).add_to(Map)


# Add a layer control panel to the map to toggle layer visibility.
Map.add_child(folium.LayerControl())


# Display the map.
display(Map)


# # Statistics Calculation and Table Generation
#
# This section calculates statistics for NDVI and NDWI within each farm polygon
# using the median composites and presents them in a downloadable table.
#

print("\n--- Statistics for Each Farm Polygon ---") # Keep this print

# Assuming the date range (start_date, end_date) and farmPolygons are defined in previous sections.
# Assuming the sentinel_ndvi_composite, sentinel_ndwi_composite, landsat_ndvi_composite,
# and landsat_ndwi_composite are available from the previous processing section.
# Assuming the clipped_srtm is available from the SRTM DEM download section.

# --- Iterate through each farm polygon and compute statistics from Composites ---
statistics_data = []

if farmPolygons is not None:
    polygons_list = farmPolygons.toList(farmPolygons.size().getInfo())

    for i in range(polygons_list.size().getInfo()):
        polygon = ee.Feature(polygons_list.get(i))
        # Attempt to get 'FID' or use a default if not available
        try:
            polygon_id = polygon.get('FID').getInfo() # Use 'FID' as the unique identifier
            if polygon_id is None:
                 polygon_id = f"Polygon_{i+1}" # Use a default if FID is None
        except Exception:
             polygon_id = f"Polygon_{i+1}" # Use a default if 'FID' property doesn't exist


        stats_row = {'Polygon ID': polygon_id}

        # Get Sentinel NDVI statistics from composite
        if sentinel_ndvi_composite is not None:
            try:
                sentinel_ndvi_stats_polygon = sentinel_ndvi_composite.select('NDVI').reduceRegion(
                    reducer=ee.Reducer.min().combine(ee.Reducer.mean(), None, True).combine(ee.Reducer.max(), None, True),
                    geometry=polygon.geometry(),
                    scale=10, # Use Sentinel-2 resolution
                    maxPixels=1e9
                )
                stats_info = sentinel_ndvi_stats_polygon.getInfo()
                if stats_info and 'min' in stats_info and stats_info.get('min') is not None:
                    stats_row['Sentinel_NDVI_Min'] = stats_info.get('min')
                    stats_row['Sentinel_NDVI_Mean'] = stats_info.get('mean')
                    stats_row['Sentinel_NDVI_Max'] = stats_info.get('max')
                else:
                    stats_row['Sentinel_NDVI_Min'] = 'N/A'
                    stats_row['Sentinel_NDVI_Mean'] = 'N/A'
                    stats_row['Sentinel_NDVI_Max'] = 'N/A'
            except Exception as e:
                stats_row['Sentinel_NDVI_Min'] = 'Error'
                stats_row['Sentinel_NDVI_Mean'] = 'Error'
                stats_row['Sentinel_NDVI_Max'] = 'Error'
        else:
            stats_row['Sentinel_NDVI_Min'] = 'N/A'
            stats_row['Sentinel_NDVI_Mean'] = 'N/A'
            stats_row['Sentinel_NDVI_Max'] = 'N/A'


        # Get Sentinel NDWI statistics from composite
        if sentinel_ndwi_composite is not None:
            try:
                sentinel_ndwi_stats_polygon = sentinel_ndwi_composite.select('NDWI').reduceRegion(
                    reducer=ee.Reducer.min().combine(ee.Reducer.mean(), None, True).combine(ee.Reducer.max(), None, True),
                    geometry=polygon.geometry(),
                    scale=10, # Use Sentinel-2 resolution
                    maxPixels=1e9
                )
                stats_info = sentinel_ndwi_stats_polygon.getInfo()
                if stats_info and 'min' in stats_info and stats_info.get('min') is not None:
                    stats_row['Sentinel_NDWI_Min'] = stats_info.get('min')
                    stats_row['Sentinel_NDWI_Mean'] = stats_info.get('mean')
                    stats_row['Sentinel_NDWI_Max'] = stats_info.get('max')
                else:
                    stats_row['Sentinel_NDWI_Min'] = 'N/A'
                    stats_row['Sentinel_NDWI_Mean'] = 'N/A'
                    stats_row['Sentinel_NDWI_Max'] = 'N/A'
            except Exception as e:
                stats_row['Sentinel_NDWI_Min'] = 'Error'
                stats_row['Sentinel_NDWI_Mean'] = 'Error'
                stats_row['Sentinel_NDWI_Max'] = 'Error'
        else:
            stats_row['Sentinel_NDWI_Min'] = 'N/A'
            stats_row['Sentinel_NDWI_Mean'] = 'N/A'
            stats_row['Sentinel_NDWI_Max'] = 'N/A'


        # Get Landsat NDVI statistics from composite
        if landsat_ndvi_composite is not None:
            try:
                landsat_ndvi_stats_polygon = landsat_ndvi_composite.select('NDVI').reduceRegion(
                    reducer=ee.Reducer.min().combine(ee.Reducer.mean(), None, True).combine(ee.Reducer.max(), None, True),
                    geometry=polygon.geometry(),
                    scale=30, # Use Landsat resolution
                    maxPixels=1e9
                )
                stats_info = landsat_ndvi_stats_polygon.getInfo()
                if stats_info and 'min' in stats_info and stats_info.get('min') is not None:
                    stats_row['Landsat_NDVI_Min'] = stats_info.get('min')
                    stats_row['Landsat_NDVI_Mean'] = stats_info.get('mean')
                    stats_row['Landsat_NDVI_Max'] = stats_info.get('max')
                else:
                    stats_row['Landsat_NDVI_Min'] = 'N/A'
                    stats_row['Landsat_NDVI_Mean'] = 'N/A'
                    stats_row['Landsat_NDVI_Max'] = 'N/A'
            except Exception as e:
                 stats_row['Landsat_NDVI_Min'] = 'Error'
                 stats_row['Landsat_NDVI_Mean'] = 'Error'
                 stats_row['Landsat_NDVI_Max'] = 'Error'
        else:
            stats_row['Landsat_NDVI_Min'] = 'N/A'
            stats_row['Landsat_NDVI_Mean'] = 'N/A'
            stats_row['Landsat_NDVI_Max'] = 'N/A'


        # Get Landsat NDWI statistics from composite
        if landsat_ndwi_composite is not None:
            try:
                landsat_ndwi_stats_polygon = landsat_ndwi_composite.select('NDWI').reduceRegion(
                    reducer=ee.Reducer.min().combine(ee.Reducer.mean(), None, True).combine(ee.Reducer.max(), None, True),
                    geometry=polygon.geometry(),
                    scale=30, # Use Landsat resolution
                    maxPixels=1e9
                )
                stats_info = landsat_ndwi_stats_polygon.getInfo()
                if stats_info and 'min' in stats_info and stats_info.get('min') is not None:
                    stats_row['Landsat_NDWI_Min'] = stats_info.get('min')
                    stats_row['Landsat_NDWI_Mean'] = stats_info.get('mean')
                    stats_row['Landsat_NDWI_Max'] = stats_info.get('max')
                else:
                    stats_row['Landsat_NDWI_Min'] = 'N/A'
                    stats_row['Landsat_NDWI_Mean'] = 'N/A'
                    stats_row['Landsat_NDWI_Max'] = 'N/A'
            except Exception as e:
                stats_row['Landsat_NDWI_Min'] = 'Error'
                stats_row['Landsat_NDWI_Mean'] = 'Error'
                stats_row['Landsat_NDWI_Max'] = 'Error'
        else:
            stats_row['Landsat_NDWI_Min'] = 'N/A'
            stats_row['Landsat_NDWI_Mean'] = 'N/A'
            stats_row['Landsat_NDWI_Max'] = 'N/A'

        # Get Mean Elevation from clipped SRTM DEM
        if 'clipped_srtm' in locals() and clipped_srtm is not None:
            try:
                # Reduce the clipped SRTM DEM image by the current polygon's geometry
                mean_elevation_stats = clipped_srtm.reduceRegion(
                    reducer=ee.Reducer.mean(),
                    geometry=polygon.geometry(),
                    scale=30, # SRTM resolution
                    maxPixels=1e9
                )
                mean_elevation_info = mean_elevation_stats.getInfo()
                # The result of reduceRegion for a single band image will be a dictionary with the band name as key
                if mean_elevation_info and 'elevation' in mean_elevation_info and mean_elevation_info.get('elevation') is not None:
                    stats_row['Mean_Elevation'] = mean_elevation_info.get('elevation')
                else:
                    stats_row['Mean_Elevation'] = 'N/A'
            except Exception as e:
                stats_row['Mean_Elevation'] = 'Error'
        else:
            stats_row['Mean_Elevation'] = 'N/A'


        statistics_data.append(stats_row)

    # Create a pandas DataFrame
    stats_df = pd.DataFrame(statistics_data)

    # Display the DataFrame
    display(stats_df)

    # Provide a link to download the DataFrame as a CSV
    # from google.colab import files # Keep this import, but ensure it's in the consolidated imports cell.
    print("\nClick the link below to download the statistics as a CSV file:")
    csv_filename = 'farm_polygon_statistics.csv'
    stats_df.to_csv(csv_filename, index=False)
    print(f"'{csv_filename}' has been created in the Colab environment.")
    print("To download this file, go to the 'Files' tab on the left sidebar, find the file, right-click, and select 'Download selection'.")

else:
    print("Farm polygons not available to generate statistics table.")


# # SRTM DEM Download
#
# This section downloads the SRTM Digital Elevation Model (DEM) for the farm polygons.
#

# Load the SRTM Digital Elevation Model data.
# Dataset ID: USGS/SRTMGL1_003 (SRTM v3 Digital Elevation Data 30m)
srtm = ee.Image('USGS/SRTMGL1_003')

# Clip the DEM to the farm polygon boundaries.
# The farmPolygons variable is assumed to be defined in a previous section.
if farmPolygons is not None:
    clipped_srtm = srtm.clip(farmPolygons)

    # Define export parameters for the clipped DEM.
    # Specify the Google Drive folder, filename prefix, region, scale, and CRS.
    # The local path C:\Roja A\Exploring\Sensegrass\Output is not directly usable for export from Colab.
    # Exporting to Google Drive is the recommended approach.
    srtm_export_task = ee.batch.Export.image.toDrive(
        image=clipped_srtm,
        description='Clipped_SRTM_DEM_FarmPolygons',
        folder='EarthEngine_Exports', # Specify a folder in your Google Drive
        fileNamePrefix='clipped_srtm_dem',
        region=farmPolygons.geometry(),
        scale=30, # SRTM resolution is approximately 30 meters
        crs='EPSG:4326', # Coordinate reference system
        maxPixels=1e9 # Increase maxPixels for potentially large exports
    )

    # Start the export task.
    ee.batch.Task.start(srtm_export_task)

    print("\nExporting clipped SRTM DEM to Google Drive...")
    print("SRTM DEM export task has been initiated to Google Drive.")
    print("You can monitor the task progress in the 'Tasks' tab in the Earth Engine Code Editor.")

else:
    print("Farm polygons not available for SRTM DEM clipping and export.")
```
